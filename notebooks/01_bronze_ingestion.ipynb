```python
# ======================================
# Imports
# ======================================
from pyspark.sql import functions as F
from pyspark.sql.types import *
```

```python
# ======================================
# Configurations
# ======================================

# Volume containing raw data
RAW_DATA_PATH = '/Volumes/retailsphere/bronze/raw_data'

# Catalog & Schemas
CATALOG = 'retailsphere'

BRONZE_SCHEMA = 'bronze'

SILVER_SCHEMA = 'silver'

GOLD_SCHEMA = 'gold'
```

```python
# ======================================
# Helper Functions
# ======================================
def generate_structtype(df, schema_name):
    print(f"{schema_name} = StructType([")
    for field in df.schema.fields:
        dtype = type(field.dataType).__name__
        nullable = "True" if field.nullable else "False"
        print(
            f'    StructField("{field.name}", {dtype}(), {nullable}),'
        )
    print("])")

customers_df = spark.read.option("header", True).option('inferSchema', True).csv('/Volumes/retailsphere/bronze/raw_data/olist_customers_dataset.csv')
generate_structtype(customers_df, "customers_schema")

geolocation_df = spark.read.option("header", True).option('inferSchema', True).csv('/Volumes/retailsphere/bronze/raw_data/olist_geolocation_dataset.csv')
generate_structtype(geolocation_df, "geolocation_schema")

orders_df = spark.read.option("header", True).option('inferSchema', True).csv('/Volumes/retailsphere/bronze/raw_data/olist_orders_dataset.csv')
generate_structtype(orders_df, "orders_schema")

order_items_df = spark.read.option("header", True).option('inferSchema', True).csv('/Volumes/retailsphere/bronze/raw_data/olist_order_items_dataset.csv')
generate_structtype(order_items_df, "order_items_schema")

order_payments_df = spark.read.option("header", True).option('inferSchema', True).csv('/Volumes/retailsphere/bronze/raw_data/olist_order_payments_dataset.csv')
generate_structtype(order_payments_df, 'order_payments_schema')

order_reviews_df = spark.read.option("header", True).option('inferSchema', True).csv('/Volumes/retailsphere/bronze/raw_data/olist_order_reviews_dataset.csv')
generate_structtype(order_reviews_df, "order_reviews_schema")

products_df = spark.read.option("header", True).option('inferSchema', True).csv('/Volumes/retailsphere/bronze/raw_data/olist_products_dataset.csv')
generate_structtype(products_df, "products_schema")

sellers_df = spark.read.option("header", True).option('inferSchema', True).csv('/Volumes/retailsphere/bronze/raw_data/olist_sellers_dataset.csv')
generate_structtype(sellers_df, "sellers_schema")

product_category_translation_df = spark.read.option("header", True).option('inferSchema', True).csv('/Volumes/retailsphere/bronze/raw_data/product_category_name_translation.csv')
generate_structtype(product_category_translation_df, "product_category_translation_schema")
```

```python
# ======================================
# Schemas
# ======================================
customers_schema = StructType([
    StructField("customer_id", StringType(), True),
    StructField("customer_unique_id", StringType(), True),
    StructField("customer_zip_code_prefix", IntegerType(), True),
    StructField("customer_city", StringType(), True),
    StructField("customer_state", StringType(), True),
])
geolocation_schema = StructType([
    StructField("geolocation_zip_code_prefix", IntegerType(), True),
    StructField("geolocation_lat", DoubleType(), True),
    StructField("geolocation_lng", DoubleType(), True),
    StructField("geolocation_city", StringType(), True),
    StructField("geolocation_state", StringType(), True),
])
orders_schema = StructType([
    StructField("order_id", StringType(), True),
    StructField("customer_id", StringType(), True),
    StructField("order_status", StringType(), True),
    StructField("order_purchase_timestamp", TimestampType(), True),
    StructField("order_approved_at", TimestampType(), True),
    StructField("order_delivered_carrier_date", TimestampType(), True),
    StructField("order_delivered_customer_date", TimestampType(), True),
    StructField("order_estimated_delivery_date", TimestampType(), True),
])
order_items_schema = StructType([
    StructField("order_id", StringType(), True),
    StructField("order_item_id", IntegerType(), True),
    StructField("product_id", StringType(), True),
    StructField("seller_id", StringType(), True),
    StructField("shipping_limit_date", TimestampType(), True),
    StructField("price", DoubleType(), True),
    StructField("freight_value", DoubleType(), True),
])
order_payments_schema = StructType([
    StructField("order_id", StringType(), True),
    StructField("payment_sequential", IntegerType(), True),
    StructField("payment_type", StringType(), True),
    StructField("payment_installments", IntegerType(), True),
    StructField("payment_value", DoubleType(), True),
])
order_reviews_schema = StructType([
    StructField("review_id", StringType(), True),
    StructField("order_id", StringType(), True),
    StructField("review_score", StringType(), True),
    StructField("review_comment_title", StringType(), True),
    StructField("review_comment_message", StringType(), True),
    StructField("review_creation_date", StringType(), True),
    StructField("review_answer_timestamp", StringType(), True),
])
products_schema = StructType([
    StructField("product_id", StringType(), True),
    StructField("product_category_name", StringType(), True),
    StructField("product_name_lenght", IntegerType(), True),
    StructField("product_description_lenght", IntegerType(), True),
    StructField("product_photos_qty", IntegerType(), True),
    StructField("product_weight_g", IntegerType(), True),
    StructField("product_length_cm", IntegerType(), True),
    StructField("product_height_cm", IntegerType(), True),
    StructField("product_width_cm", IntegerType(), True),
])
sellers_schema = StructType([
    StructField("seller_id", StringType(), True),
    StructField("seller_zip_code_prefix", IntegerType(), True),
    StructField("seller_city", StringType(), True),
    StructField("seller_state", StringType(), True),
])
product_category_translation_schema = StructType([
    StructField("product_category_name", StringType(), True),
    StructField("product_category_name_english", StringType(), True),
])
```

```python
# ======================================
# Ingestion Function
# ======================================

def ingest_to_bronze(
    filepath: str,
    dataframe_schema: StructType,
    catalog: str,
    database: str,
    table_name: str,
    mode: str = "append"
):
    # Read CSV
    df = spark.read.option("header", True).schema(dataframe_schema).csv(filepath)

    # Add metadata columns
    df = df.withColumns({
        "ingestion_timestamp": F.current_timestamp(),
        "source_file": F.col('_metadata.file_path')
    })

    # Count rows (triggers action)
    try:
        row_count = df.count()

        # Write Delta table
        df.write.format("delta").mode(mode).saveAsTable(f"{catalog}.{database}.{table_name}")

        print(f"""
=========================================
Bronze Ingestion Completed
=========================================
Table      : {table_name}
Rows       : {row_count}
Source     : {filepath}
Target     : {catalog}.{database}.{table_name}
=========================================
""")

        return {
            "table": table_name,
            "rows": row_count
        }

    except Exception as e:
        print(f"""
=========================================
Bronze Ingestion Failed
=========================================
Table  : {table_name}
Source : {filepath}

Error:
{e}
=========================================
""")
        return None
```

```python
# ======================================
# Ingesting Bronze Tables
# ======================================
datasets = [

    {
        "table": "customers",
        "path": f"{RAW_DATA_PATH}/olist_customers_dataset.csv",
        "schema": customers_schema
    },

    {
        "table": "geolocation",
        "path": f"{RAW_DATA_PATH}/olist_geolocation_dataset.csv",
        "schema": geolocation_schema
    },

    {
        "table": "order_items",
        "path": f"{RAW_DATA_PATH}/olist_order_items_dataset.csv",
        "schema": order_items_schema
    },

    {
        "table": "order_payments",
        "path": f"{RAW_DATA_PATH}/olist_order_payments_dataset.csv",
        "schema": order_payments_schema
    },

    {
        "table": "order_reviews",
        "path": f"{RAW_DATA_PATH}/olist_order_reviews_dataset.csv",
        "schema": order_reviews_schema
    },

    {
        "table": "orders",
        "path": f"{RAW_DATA_PATH}/olist_orders_dataset.csv",
        "schema": orders_schema
    },

    {
        "table": "products",
        "path": f"{RAW_DATA_PATH}/olist_products_dataset.csv",
        "schema": products_schema
    },

    {
        "table": "sellers",
        "path": f"{RAW_DATA_PATH}/olist_sellers_dataset.csv",
        "schema": sellers_schema
    },

    {
        "table": "product_category_translation",
        "path": f"{RAW_DATA_PATH}/product_category_name_translation.csv",
        "schema": product_category_translation_schema
    }

]

results = []

for dataset in datasets:

    result = ingest_to_bronze(
        filepath=dataset["path"],
        dataframe_schema=dataset["schema"],
        catalog=CATALOG,
        database=BRONZE_SCHEMA,
        table_name=dataset["table"],
        mode = "overwrite"
    )

    if result:
        results.append(result)


print("=" * 40)
print("Bronze Layer Summary")
print("=" * 40)

total_rows = 0

for result in results:
    print(f"{result['table']:<30} {result['rows']:>10}")
    total_rows += result["rows"]

print("-" * 40)
print(f"Total Rows Loaded : {total_rows}")
print("=" * 40)
```
